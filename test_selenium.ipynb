```python
url = "https://www.hapag-lloyd.com/en/online-business/schedules/vessel-tracing.html"
driver.get(url)
```

```python
#driver.find_element_by_id("schedules_vessel_tracing_f:hl16")
time.sleep(10)

element = driver.find_element_by_id("schedules_vessel_tracing_f:hl16")
all_options = element.find_elements_by_tag_name("option")

for option in all_options:
    optValue = option.get_attribute("value")
    if not optValue:
        continue
    optText = option.text
    print("Value is: {0} = {1}".format(optValue, optText))
    time.sleep(10)
    option.click()
    break

#driver.find_element_by_id("schedules_vessel_tracing_f:hl24").click()
```
